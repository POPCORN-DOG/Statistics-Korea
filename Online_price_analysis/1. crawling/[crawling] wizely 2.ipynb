```python
import requests
import json
import pandas as pd
import warnings

from pandas import DataFrame

pd.set_option("display.max_rows", None , "display.max_columns", None)
warnings.filterwarnings(action='ignore')
```

```python
def wizely(count='1'):
    url = 'https://review-api.wiselycompany.com/api/1.0/review/elasticSearchGetReviewList.do'

    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.111 Safari/537.36'
    }

    data = {
        'page': '1',
        'rgid': '1',
        'perpage': str(count),
        'keywords':'' 
    }

    resp = requests.post(url,headers=headers,data=data, verify=False).text
    jj = json.loads(resp)
    return jj
```

```python
df = DataFrame(wizely(wizely()['Pagination']['TotalCount'])['ReviewList'])
df.head()
```
